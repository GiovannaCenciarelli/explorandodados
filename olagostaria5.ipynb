```python
from PIL import Image, ImageDraw, ImageFont
import imageio

def create_funny_image_with_gif(gif_path):
    # Defina as dimenses da imagem
    width, height = 400, 200
    
    # Criar uma lista para armazenar os quadros do GIF
    gif_with_text = []

    # Criar quadros para o GIF
    for i in range(10):
        # Criar uma nova imagem com fundo branco
        img = Image.new('RGB', (width, height), color='white')
        draw = ImageDraw.Draw(img)
        
        # Adicionar o texto
        text = "Quando voc est 'unbothered' escrevendo seu cdigo "
        font = ImageFont.load_default()
        
        # Obter dimenses do texto
        bbox = draw.textbbox((0, 0), text, font=font)
        text_width = bbox[2] - bbox[0]
        text_height = bbox[3] - bbox[1]
        
        # Centralizar o texto
        draw.text(((width - text_width) / 2, (height - text_height) / 2), text, fill='black', font=font)
        
        # Adicionar a imagem  lista de quadros
        gif_with_text.append(img)

    # Salvar o GIF
    imageio.mimsave(gif_path, gif_with_text, duration=0.1)

# Caminho do GIF a ser salvo
gif_path = 'funny_gif_with_text.gif'
create_funny_image_with_gif(gif_path)

print("GIF gerado com sucesso: 'funny_gif_with_text.gif'")
```
